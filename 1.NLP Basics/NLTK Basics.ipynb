```python
import pandas as pd
import nltk
import urllib.request
from bs4 import BeautifulSoup
```

## Read HTML Content from the http://php.net/

```python
response = urllib.request.urlopen('http://php.net/')
```

```python
html=response.read()
```

```python
print(html)
```

## Format the HTML Content

```python
soup = BeautifulSoup(html,"html5lib")
```

```python
soup
```

## Get Plain Text from the HTML Content

```python
text = soup.get_text(strip=True)
```

```python
print(text)
```

## Split every word form the given content and push into the array 

```python
token=[t for t in text.split()]
```

```python
print(token)
```

```python
freq = nltk.FreqDist(token)
```

```python
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
from nltk.corpus import stopwords
st=stopwords.words('english')
```

```python
clean_tokens = token[:]
```

```python
clean_tokens = [t.lower() for t in clean_tokens]
```

```python
for token in token:
    if token in st:
        clean_tokens.remove(token)
```

```python
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
freq.plot(30,cumulative=False)
```
